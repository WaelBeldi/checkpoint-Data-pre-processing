```python
# 4. Inspect the dataset for potential missing values.
df.isnull().sum()
```

```python
# 5. Select your strategy to handle missing values, and tell us why you had made that choice.
# Handle 'counter_number' null values: Change type from float to string then fill with 'Unknown' for unknown counter numbers.
df['counter_number'] = df['counter_number'].astype(str)
df['counter_number'].fillna(value='Unknown', inplace=True)

# Handle 'reading_remarque' null values: Replace with median because the data distribution is skewed.
df['reading_remarque'] = df['reading_remarque'].fillna(df['reading_remarque'].median())

df.isnull().sum()
```

```python
# 8. Transform the 'counter_type' feature to a numeric variable using the encoder of your choice.
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['counter_type'] = encoder.fit_transform(df['counter_type'])
df
```
